```python

df_dois = pd.DataFrame(df['DIRETOR'].value_counts())
df_tres = df_dois.assign(diretor=df_dois.index)
df_tres.reset_index(drop=True)
df_tres.info()
```

```python
import matplotlib.pyplot as plt
```

```python
df_pie = df_tres['count'].head(10)
labels = df_tres['diretor'].head(10)

# df_pie = df_filme['TOTAL_TECNCIOS_ESTRANGEIROS']


fig, ax = plt.subplots()
ax.pie(df_pie, labels=labels,
       shadow=True, startangle=90)

# ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
#        ylim=(0, 8), yticks=np.arange(1, 8))

plt.title('Co-produções Estrangeiras no Brasil | 2020 - 2021')
```
