```python
import pandas as pd
import numpy as np
```

```python
df_invest_renun_fiscal = pd.read_csv('InvestidoresEmProjetosRenunciaFiscal.csv', encoding='latin1', sep=';')
df_invest_renun_fiscal
```

```python

cols_to_convert = ['ART25', 'ART18', 'ART1A', 'ART1', 'ART39', 'ART3A', 'ART3', 'ART41', 'VL_TOTAL_INVESTIDO']

for col in cols_to_convert:
    if df_invest_renun_fiscal[col].dtype == object:
        df_invest_renun_fiscal[col] = df_invest_renun_fiscal[col].str.replace(',', '.').astype(float)

df_invest_renun_fiscal

df_invest_renun_fiscal['PESSOA_FISICA'] = df_invest_renun_fiscal['CNPJ_INVESTIDOR'].apply(lambda x: True if x == 'Pessoa Fisica' else False)
df_invest_renun_fiscal['CNPJ'] = df_invest_renun_fiscal['CNPJ_INVESTIDOR'].apply(lambda x: True if x != 'Pessoa Fisica' else False)

df_invest_renun_fiscal
```

```python
df_invest_renun_fiscal_cnpj = df_invest_renun_fiscal[df_invest_renun_fiscal['CNPJ_INVESTIDOR'].str.contains(r'\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}')]
df_invest_renun_fiscal_cpf = df_invest_renun_fiscal[df_invest_renun_fiscal['CNPJ_INVESTIDOR'] == 'Pessoa Fisica']

```

```python
top_15_investors = df_invest_renun_fiscal_cnpj.nlargest(20, 'VL_TOTAL_INVESTIDO')

print("Top 15 investors with the highest VL_TOTAL_INVESTIDO:")
print(top_15_investors[['INVESTIDOR', 'VL_TOTAL_INVESTIDO']])
```
