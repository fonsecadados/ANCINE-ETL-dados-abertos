```python
import pandas as pd
import numpy as np
```

```python
df_obras_n_publi_br = pd.read_csv('ObrasNaoPublicitariasBrasileiras.csv', encoding='latin1', sep=';')
df_obras_n_publi_br = df_obras_n_publi_br[['TITULO_ORIGINAL', 'REQUERENTE', 'TIPO_OBRA', 'ORGANIZACAO_TEMPORAL', 'ANO_PRODUCAO', 'DT_EMISSAO_CPB', 'COPRODUCAO_INTERNACIONAL', 'FOMENTO_INDIRETO', 'INVESTIMENTO_FSA']]
df_obras_n_publi_br.info()
```

```python
df_obras_n_publi_br
```

```python
unique_values = df_obras_n_publi_br['DT_EMISSAO_CPB'].unique()
unique_values
```

```python

df_obras_n_publi_br['COPRODUCAO_INTERNACIONAL'] = df_obras_n_publi_br['COPRODUCAO_INTERNACIONAL'].replace(['NÃO', 'SIM'],[0,1]) #Método com replace
df_obras_n_publi_br['FOMENTO_INDIRETO'] = df_obras_n_publi_br['FOMENTO_INDIRETO'].replace(['NÃO', 'SIM'],[0,1]) #Método com replace
df_obras_n_publi_br['INVESTIMENTO_FSA'] = df_obras_n_publi_br['INVESTIMENTO_FSA'].replace(['NÃO', 'SIM'],[0,1]) #Método com replace

df_obras_n_publi_br['DT_EMISSAO_CPB'] = pd.to_datetime(df_obras_n_publi_br['DT_EMISSAO_CPB'], format='mixed', dayfirst=True)
df_obras_n_publi_br['DT_EMISSAO_CPB'] = df_obras_n_publi_br['DT_EMISSAO_CPB'].dt.year

df_obras_n_publi_br['ANO_PRODUCAO'].sort_values().head(25)

df_obras_n_publi_br.to_csv('obras_nao_publi_br.csv')

```
