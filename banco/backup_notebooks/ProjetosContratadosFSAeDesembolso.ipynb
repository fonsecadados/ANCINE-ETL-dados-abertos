```python
for index, row in df_prj_fsa_dembolso.iterrows():
    if pd.isnull(row['TITULO_PROJETO']):
        df_prj_fsa_dembolso.at[index, 'TITULO_PROJETO'] = row['CHAMADA_PUBLICA']

for index, row in df_prj_fsa_dembolso.iterrows():
    if pd.isnull(row['RAZAO_SOCIAL_PRODUTORA']):
        df_prj_fsa_dembolso.at[index, 'RAZAO_SOCIAL_PRODUTORA'] = row['RAZAO_SOCIAL_PROPONENTE']

for index, row in df_prj_fsa_dembolso.iterrows():
    if pd.isnull(row['RAZAO_SOCIAL_DISTRIBUIDORA_PROGRAMADORA']):
        df_prj_fsa_dembolso.at[index, 'RAZAO_SOCIAL_DISTRIBUIDORA_PROGRAMADORA'] = 'SEM DADOS'

for index, row in df_prj_fsa_dembolso.iterrows():
    if pd.isnull(row['VL_TOTAL_LIBERADO']):
        df_prj_fsa_dembolso.at[index, 'VL_TOTAL_LIBERADO'] = 0


for index, row in df_prj_fsa_dembolso.iterrows():
    if pd.isnull(row['DT_PRIMEIRO_DESEMBOLSO']):
        df_prj_fsa_dembolso.at[index, 'DT_PRIMEIRO_DESEMBOLSO'] = 0



df_prj_fsa_dembolso.isnull().sum()
```

```python
df_prj_fsa_dembolso['ANO_CHAMADA_PUBLICA'] = df_prj_fsa_dembolso['ANO_CHAMADA_PUBLICA'].str.extract(r'(\d{4})')
df_prj_fsa_dembolso['ANO_CHAMADA_PUBLICA'] = pd.to_numeric(df_prj_fsa_dembolso['ANO_CHAMADA_PUBLICA'])
df_prj_fsa_dembolso['DT_PUBLICACAO_CONTRATO_DOU'] = pd.to_datetime(df_prj_fsa_dembolso['DT_PUBLICACAO_CONTRATO_DOU'], format='%d/%m/%Y', dayfirst=True)
```
