```python
import pandas as pd
import numpy as np
```

```python
df_salas_exib = pd.read_csv('SalasDeExibicaoEComplexos.csv', encoding='latin1', sep=';')
df_salas_exib = df_salas_exib[['NOME_SALA', 'SITUACAO_SALA',
       'ASSENTOS_SALA', 'NOME_COMPLEXO',
       'MUNICIPIO_COMPLEXO', 'UF_COMPLEXO',
       'COMPLEXO_ITINERANTE', 'OPERACAO_USUAL', 'SITUACAO_EXIBIDOR',
       'GRUPO_EXIBIDOR']]
```

```python
df_salas_exib
```

```python
df_salas_exib.isnull().sum()
```

```python
for index, row in df_salas_exib.iterrows():
    if pd.isnull(row['ASSENTOS_SALA']):
        df_salas_exib.at[index, 'ASSENTOS_SALA'] = 0


for index, row in df_salas_exib.iterrows():
    if pd.isnull(row['NOME_SALA']):
        df_salas_exib.at[index, 'NOME_SALA'] = 'SEM NOME'


for index, row in df_salas_exib.iterrows():
    if pd.isnull(row['MUNICIPIO_COMPLEXO']):
        df_salas_exib.at[index, 'MUNICIPIO_COMPLEXO'] = 'SEM UF'


for index, row in df_salas_exib.iterrows():
    if pd.isnull(row['UF_COMPLEXO']):
        df_salas_exib.at[index, 'UF_COMPLEXO'] = 'SEM UF'


for index, row in df_salas_exib.iterrows():
    if pd.isnull(row['SITUACAO_SALA']):
        df_salas_exib.at[index, 'SITUACAO_SALA'] = 'SEM STATUS'
```

```python
df_salas_exib['ASSENTOS_SALA'] = df_salas_exib['ASSENTOS_SALA'].astype(int)
```

```python
df_salas_exib.to_csv('salas_exibi.csv')
```
