```python
import pandas as pd
import numpy as np
```

```python
df_prg_apoio_mercado = pd.read_csv('ProgramaApoioAMercados.csv', encoding='latin1', sep=';')
df_prg_apoio_mercado = df_prg_apoio_mercado[['NOME_EVENTO', 'DT_INICIO_EVENTO', 'RAZAO_SOCIAL_PRODUTORA', 'REPRESENTANTE', 'VALOR_APOIO_FINANCEIRO']]
df_prg_apoio_mercado['DT_INICIO_EVENTO'] = pd.to_datetime(df_prg_apoio_mercado['DT_INICIO_EVENTO'], format='mixed', dayfirst=True)
df_prg_apoio_mercado['DT_INICIO_EVENTO'] = df_prg_apoio_mercado['DT_INICIO_EVENTO'].dt.year

df_prg_apoio_mercado['VALOR_APOIO_FINANCEIRO'] = df_prg_apoio_mercado['VALOR_APOIO_FINANCEIRO'].apply(lambda x: float(x.replace(',', '.')))
df_prg_apoio_mercado
```

```python
df_prg_apoio_mercado.to_csv('prg_apoio_mercado.csv')
```
