```python
import pandas as pd
import numpy as np
```

```python
df_apoio= pd.read_csv('ApoioFestivaisLaboratoriosWorkshops.csv', encoding='latin1', sep=';')
df_apoio['DT_INICIO_FESTIVAL']= pd.to_datetime(df_apoio['DT_INICIO_FESTIVAL'], format='%d/%m/%Y', dayfirst=True)
df_apoio_fest = df_apoio[['FESTIVAL', 'TITULO_ORIGINAL', 'TIPO_APOIO', 'METRAGEM','REPRESENTANTE', 'RAZAO_SOCIAL_PRODUTORA', 'VALOR_APOIO']]
df_apoio_fest['ANO'] = df_apoio['DT_INICIO_FESTIVAL'].dt.year
df_apoio_fest['VALOR_APOIO'] = df_apoio_fest['VALOR_APOIO'].apply(lambda x: float(x.replace(',', '.')))
df_apoio_fest
```

```python
def fill_nan(column, row, new_value):
    df_apoio_fest.at[row, column] = new_value

fill_nan('RAZAO_SOCIAL_PRODUTORA', 4, 'Daniel Lentini Faria')
fill_nan('RAZAO_SOCIAL_PRODUTORA', 19, 'Maria Julia Pereira de Paiva')
fill_nan('RAZAO_SOCIAL_PRODUTORA', 29, 'Letícia Fudissaku')
fill_nan('RAZAO_SOCIAL_PRODUTORA', 39, 'Mariana Jaspe de Souza Oliveira')
fill_nan('METRAGEM', 83, 'Projeto de Longa')
```
