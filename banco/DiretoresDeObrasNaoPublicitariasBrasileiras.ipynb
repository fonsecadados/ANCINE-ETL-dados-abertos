```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DiretoresDeObrasNaoPublicitariasBrasileiras.csv', encoding='latin1', sep=';')
df_clean = df[['TITULO_ORIGINAL', 'DIRETOR']]
df_dir_sum = df_clean['DIRETOR'].value_counts().nlargest(15)
```

```python
df_dir = pd.DataFrame(df_dir_sum)
df_dir_br_publi = df_dir.reset_index()
df_dir_br_publi.columns = ['DIRETOR', 'TOTAL']
df_dir_br_publi
df_dir_br_publi.to_csv('diretores_publi.csv', index=False)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

count = df_dir_br_publi['TOTAL'][::-1]
df_dir2 = df_dir_br_publi['DIRETOR'][::-1]
bars = ax.barh(df_dir2, count)

ax.set_title('Agentes Econmicos ANCINE - ES')
ax.spines[['right', 'top', 'bottom']].set_visible(False)
ax.xaxis.set_visible(False)


ax.bar_label(ax.containers[0])
ax.set_xlim(left=4)


plt.show()
```
