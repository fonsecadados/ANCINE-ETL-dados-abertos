```python
import pandas as pd
import numpy as np
```

```python
df_canais_regis = pd.read_csv('CanaisDasProgramadorasRegistradas.csv', encoding='latin1', sep=';')
df_canais_regis_filt =  df_canais_regis[['CANAL', 'CLASSIFICACAO_CANAL', 'TIPO_CONTEUDO_CANAL', 'DATA_INICIO_OFERTA', 'DENSIDADE_CANAL', 'PAIS_PROGRAMADORA']]
df_canais_regis_filt
```

```python
df_canais_regis_filt['DATA_INICIO_OFERTA'] = pd.to_datetime(df_canais_regis_filt['DATA_INICIO_OFERTA'], format='%d/%m/%Y', dayfirst=True)
df_canais_regis_filt['DATA_INICIO_OFERTA'] = df_canais_regis_filt['DATA_INICIO_OFERTA'].dt.year
df_canais_regis_filt
```
